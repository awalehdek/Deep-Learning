# Logistic Regression

Given x, we want ŷ = P(y = 1/x) 

P(y = 1/x) is the probability of y=1 (chicken) given input x picture
P(y = 1/x) in another word if x is a picture waht is the chance of x to be a chicken.

x ∈ ℝ^(Nx)

Parameters: w ∈ ℝ^(Nx), b ∈ ℝ

Output : ŷ = w^(T)x + b   convert to sigmoid function : ŷ = σ (w^(T)x + b) / z

if sigmoid function : σ(z)  =1 /(1+e^(-z)).  

if z is large σ(z) ≈ 0 

if z is small  σ(z) ≈ 1
